## Yearly Analysis of Traffic Accident Causes

```python
# Pivot the DataFrame
df_grouped_by_year2 = df_grouped_by_year.pivot(index='JAHR', columns='MONATSZAHL', values='WERT')

# Calculate percentages
totals = df_grouped_by_year2.sum(axis=1)
percentages = df_grouped_by_year2.div(totals, axis=0) * 100

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(14, 12), sharey=True)
boroughs = list(df_grouped_by_year2.columns)
colors = ['rocket', 'OrRd', 'BuPu']


for i, (borough, cmap) in enumerate(zip(boroughs, colors)):
    norm = plt.Normalize(percentages[borough].min(), percentages[borough].max())
    sm = cm.ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])
    
    bars = axes[i].barh(df_grouped_by_year2.index, df_grouped_by_year2[borough], 
                        color=sm.to_rgba(percentages[borough]))
    axes[i].set_title(borough)
    axes[i].set_xlim(0, df_grouped_by_year2.max().max() * 1.1)
    axes[i].invert_yaxis()
    if i == 0:
        axes[i].set_ylabel('Years')
    
    # Add percentage labels
    for bar, percentage in zip(bars, percentages[borough]):
        width = bar.get_width()
        axes[i].text(width, bar.get_y() + bar.get_height()/2, f'{percentage:.1f}%', 
                     va='center', ha='left', fontsize=10)
    
    # Add colorbar
    cbar = fig.colorbar(sm, ax=axes[i], orientation='vertical', pad=0.01)
    cbar.set_label('Percentage of Accidents')

fig.suptitle('Yearly Analysis of Traffic Accident Causes', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()
```

## Comparison of "Overall", "With personal Injury" and "Injured and killed" for each Type of Accident 

```python
# Filtering data for 'insgesamt', 'mit Personenschäden' and 'Verletzte und Getötete'
filtered_df = df[df['AUSPRAEGUNG'].isin(['insgesamt','mit Personenschäden', 'Verletzte und Getötete'])]

# Grouping by 'MONATSZAHL' and 'AUSPRAEGUNG' and summing up the 'WERT'
grouped_df = filtered_df.groupby(['MONATSZAHL', 'AUSPRAEGUNG'])['WERT'].sum().unstack()

# Setting the color palette to 'rocket'
colors = sns.color_palette('rocket', n_colors=grouped_df.shape[1])

# Plotting
ax = grouped_df.plot(kind='bar', figsize=(12, 8), width=0.8, color=colors)

# Setting the labels and title
ax.set_xlabel('Type of Accident', fontweight='bold')
ax.set_ylabel('Number of Accidents')
ax.set_title('Comparison of "Overall", "With personal Injury" and "Injured and killed" for each Type of Accident')

# Setting xticks to be horizontal and bold
plt.xticks(rotation=0, ha='center', fontweight='bold')

# Adding original values on top of the bars
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.0f}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.tight_layout()
plt.legend(title='AUSPRAEGUNG')
plt.show()
```
